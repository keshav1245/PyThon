```python
def print_board(board):
    print(f'     |     |     \n {board[1]:^3} | {board[2]:^3} | {board[3]:^3} \n     |     |     ')
    print('-----------------')
    print(f'     |     |     \n {board[4]:^3} | {board[5]:^3} | {board[6]:^3} \n     |     |     ')
    print('-----------------')
    print(f'     |     |     \n {board[7]:^3} | {board[8]:^3} | {board[9]:^3} \n     |     |     ')
```

```python
playerMoves = ['#','1','2','3','4','5','6','7','8','9']
```

```python
import time
from IPython.display import clear_output

print("Welcome to Tic Tac Toe !")
p1Choice = ''
p2Choice = ''
time.sleep(0.05)
while True:
    p1Choice = input("\nPlayer 1: Do you want to be X or O? ")
    if p1Choice.upper() == 'X':
        p2Choice = 'O'
        break
    elif p1Choice.upper() == 'O':
        p2Choice = 'X'
        break
    else:
        continue
        
def checkWinner(label):
    if playerMoves[1] == playerMoves[2] == playerMoves[3] == label:
        return True
    
    if playerMoves[1] == playerMoves[4] == playerMoves[7] == label:
        return True
    
    if playerMoves[1] == playerMoves[5] == playerMoves[9] == label:
        return True
    
    if playerMoves[2] == playerMoves[5] == playerMoves[8] == label:
        return True
    
    if playerMoves[3] == playerMoves[6] == playerMoves[9] == label:
        return True
    
    if playerMoves[7] == playerMoves[5] == playerMoves[3] == label:
        return True
    
    if playerMoves[4] == playerMoves[5] == playerMoves[6] == label:
        return True
    
    if playerMoves[7] == playerMoves[8] == playerMoves[9] == label:
        return True
    
    return False
    
    

def playGame(playerMark):
    playBool = True
    index = 0
    while playBool:
        clear_output()
        print_board(playerMoves)
        pos = 0
        while True:
            pos = int(input("Choose your next Position: (1-9)"))
            #if not (playerMoves[pos] == p1Choice or playerMoves[pos] == p2Choice):
             #   continue
            if pos>=1 and pos <= 9:
                if not (playerMoves[pos] == p1Choice or playerMoves[pos] == p2Choice):
                    break
                else:
                    continue
            else:
                continue
            
        playerMoves[pos] = playerMark

        
        index += 1  
        
        if checkWinner(playerMark):
            clear_output()
            print_board(playerMoves)
            if playerMark == p1Choice:
                print("Player 1 Wins !")
                break
            else:
                print("Player 2 Wins !")
                break
        
        if index == 9:
            print("Match Draw !")
            break
        
        if playerMark == p1Choice:
            playerMark = p2Choice
            print("Player 2's Turn !")
        else:
            playerMark = p1Choice
            print("Player 1's Turn !")
            

print(f"Player 1 took {p1Choice}, thus Player 2 is {p2Choice}")
print("Player 1 will go first.")
time.sleep(0.05)
while True:
    play = input('Are you ready to play ? Enter Yes or No? ')
    if play.upper() == "YES":
        playGame(p1Choice)
        break
    elif play.upper() == "NO":
        print("See ya Later...!!!")
        break
    else:
        continue
```
